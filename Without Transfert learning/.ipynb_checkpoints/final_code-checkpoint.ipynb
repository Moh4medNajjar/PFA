Importing needed libraries : 

```python
import tensorflow as tf
from tensorflow.keras import models, layers 
import matplotlib.pyplot as plt
import os
import shutil
from PIL import Image, ImageEnhance
from tqdm import tqdm
from tensorflow.keras.utils import plot_model
import numpy as np
from sklearn.metrics import classification_report,confusion_matrix
from keras.callbacks import ReduceLROnPlateau
```

**TRANSFERRING DATASET TO MAKE 2 CLASSES**

The purpose of the following code is to reorganize the dataset into a simpler structure. The original dataset consists of three subfolders: train_set, test_set, and validation_set. Each of these subfolders contains two more subfolders: normal and pneumonia. This means that there are six subfolders in total, each with a different subset of the data. The code below merges the data from the three subfolders into two main folders: Normal and Pneumonia. This way, the dataset is easier to access and manipulate.

```python

data_dir = '/kaggle/input/chest-xray-pneumonia/chest_xray'

# Define the directory paths for the normal and abnormal data
normal_dirs = ['train/NORMAL', 'test/NORMAL', 'val/NORMAL']
abnormal_dirs = ['train/PNEUMONIA', 'test/PNEUMONIA', 'val/PNEUMONIA']

# Create a new directory to store the transformed data
os.makedirs(os.path.join('/kaggle/working/new_data'), exist_ok=True)

# Loop through the normal and abnormal directories, and copy the images to the new directory
for directory in normal_dirs + abnormal_dirs:
    src_dir = os.path.join(data_dir, directory)
    filenames = os.listdir(src_dir)
    for filename in filenames:
        src_path = os.path.join(src_dir, filename)
        dst_path = os.path.join('/kaggle/working/new_data', directory.split('/')[1], filename)
        os.makedirs(os.path.dirname(dst_path), exist_ok=True)
        shutil.copy(src_path, dst_path)
```

**DATA IMPORTATION**

Here we are defining certain constants to ensure that any subsequent alteration to them can be carried out in a single location within the codebase, thus eliminating the need to change them everywhere in the code .

```python
BATCH_SIZE = 32 
IMAGE_SIZE = 256
EPOCHS=20
CHANNELS=3
```

In the following code, we are import the dataset from Kaggle following the earlier mentionned alterations. Then, we proceed to define several variables, including the image size, batch size, and shuffle parameter.

When the "shuffle" parameter is set to "True," the data will be imported in a randomized order with each execution of the code. This approach enhances the model's robustness.

* The output will indicate the detection of two distinct classes comprising a total of 5856 images

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/kaggle/working/new_data",
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
)
```

**DATASET PARTITION**

* In the this piece of code we are defining a function taking 6 arguments .. The purpose of this function is to re_arrange the given dataset based on the percentage distribution of the training, validation, and testing datasets from the input dataset. . The resulting output of the function will be 3 variables that contain images allocated based on the specified percentage distribution.

```python
def get_dataset_partitions_tf(ds , train_split=0.8 , val_split=0.1 , test_split=0.1 , shuffle=True,shuffle_size=1000):
    ds_size = len(ds)
    if shuffle :
        ds = ds.shuffle(shuffle_size , seed = 12)
    
    train_ds = ds.take(int(ds_size*train_split))
    val_ds = ds.take(int(ds_size*val_split))
    test_ds = ds.skip(len(val_ds)+len(train_ds))
    
    
    
    return train_ds , val_ds , test_ds 
```

Calling the function here and assigning it to 3 variables ..
Now , our dataset is ready !

```python
train_ds , val_ds , test_ds = get_dataset_partitions_tf(dataset)
```

**TRAIN DATA EXAMPLES**

**DATA PREPROCESSING**

The following code we are performing three important operations on the train_ds, test_ds, and val_ds datasets (namely caching, shuffling, and pre-fetching) :

Caching involves storing the data in memory for faster access during subsequent iterations of the dataset. This can speed up the training process .

Shuffling the data ensures that the model does not learn to recognize the order of the data .

These operations can improve the efficiency of the training process, leading to faster convergence and better performance of the model.

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

Here we are creating the first layer, which involves resizing and rescaling the images :
Resising : setting the image dimensions 
We performed a grayscale normalization to reduce the effect of illumination's differences.Moreover the CNN converges faster on [0..1] data than on [0..255].

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE , IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255)
    
])
```

Here, we are creating another variable containing a layer for data augmentation that will perform random flip and random rotation on each image .
This helps to increase the variability of the training data, and improve the model's ability to generalize to new data, leading to better performance on real-world problems.

```python
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomRotation(0.2),
])
```

**CREATING THE MODEL**

In the following code , we are actually building the model :
The model consists of several layers including Conv2D, MaxPooling2D, and Dense layers. It also includes the preprocessing layers for resizing, rescaling, and data augmentation.

*The activation functions that used are :"ReLU" and "Softmax" .. We tried "Sigmoid" but the results were not good ..

```python
input_shape=(BATCH_SIZE,IMAGE_SIZE,IMAGE_SIZE,CHANNELS)
n_classes = 3
model = models.Sequential([
    resize_and_rescale,
    data_augmentation,
    layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, kernel_size = (3,3),activation='relu' ),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, kernel_size = (3,3),activation='relu' ),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3),activation='relu' ),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3),activation='relu' ),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3) ,activation='relu' ),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(n_classes,activation='softmax'),
    
])
```

Building the model by defining its input shape ..

```python
model.build(input_shape=(None,IMAGE_SIZE,IMAGE_SIZE,CHANNELS))  # Build the model
```

Next, we are compiling the model, setting up the necessary configurations for training the model such as specifying the optimizer, loss function, and evaluation metric. It is necessary to compile the model before training it. (We tried out a couple of optimizers before choosing this one as it delivered the best results)

```python
model.compile(
optimizer='adam',
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
metrics=['accuracy'])
```

```python
plot_model(model, show_shapes=True)
```

The next snippet of code creates a learning rate scheduler that reduces the learning rate when the validation accuracy stops improving remarquably .

```python
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience = 2, verbose=1,factor=0.3, min_lr=0.000001)
```

**TRAINING THE MODEL**

We are ready to go, training the model .. 

```python
history = model.fit(
    train_ds,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1,
    validation_data=val_ds,
    callbacks = [learning_rate_reduction]
    
)
```

**Evaluating the model**

Now, we will evaluate the performance of our model based on the validation dataset and storing the score in a variable .. 

```python
scores = model.evaluate(test_ds)
```

Next, we are displaying the confusion matrix in order to visually check the accuracy of our model .

```python
# Predict labels for test data
y_pred = []
for images, labels in test_ds:
    y_pred_batch = model.predict(images)
    y_pred_batch = np.argmax(y_pred_batch, axis=-1)
    y_pred.extend(y_pred_batch)

# Get true labels for test data
y_true = []
for images, labels in test_ds:
    y_true_batch = labels.numpy()
    y_true.extend(y_true_batch)

# Print the confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

# Print the classification report
cr = classification_report(y_true, y_pred, target_names=class_names)
print("Classification Report:")
print(cr)
```

Moving forward, we shall start testing our fresh model by initially displaying a randomly selected image from our testing dataset along with its corresponding label, followed by the model's prediction

```python
class_names=["Pneumonia","Normal"]
for images_batch, label_batch in test_ds.take(1):
    image = images_batch[0].numpy().astype("uint8")
    label = label_batch[0].numpy()
    
    print("first image to predict : ")
    plt.imshow(image)
    print("Actual label : ",class_names[label])
    batch_prediction = model.predict(images_batch)
    print("Predicted label : ",class_names[np.argmax(batch_prediction[0])])
```

Finally, saving the model .
